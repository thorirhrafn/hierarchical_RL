```python
import numpy as np
```

```python
from pathlib import Path

folder = Path("./dataset/")

for npz_file in folder.glob("*.npz"):
    with np.load(npz_file) as data:
        print(f"Loaded {npz_file.name}:")
        for key in data.files:
            print(f"  - {key}: shape = {data[key].shape}, dtype = {data[key].dtype}")
```
